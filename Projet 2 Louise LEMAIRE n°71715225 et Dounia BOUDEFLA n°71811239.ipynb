```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = np.linspace(-10, 10, num=50)
y_data = 0.1*np.sin(x_data) + 0.5*np.cos(x_data) + 0.2*np.random.normal(size=50)
```

```python
def f(x_data, y_data):    # Fonction qui permet de calculer la valeur de a et b 
    U = 0
    V = 0
    W = 0
    P = 0
    Q = 0
    for xi in x_data:     # On calcule les sommes U,V,W,P,Q
        U += (np.sin(0.5*np.pi*xi))**2
        V += np.cos(0.5*np.pi*xi)*np.sin(0.5*np.pi*xi)
        W += (np.cos(0.5*np.pi*xi))**2
    for xi, yi in zip(x_data, y_data):
        P += yi*np.sin(0.5*np.pi*xi)
        Q += yi*np.cos(0.5*np.pi*xi)
    if V**2-U*W == 0:
        print("Erreur, le dénominateur est nul.")
    else:
        a = (Q*V-P*W)/(V**2-U*W)
        b = (P*V-Q*U)/(V**2-U*W)
    return a, b

a,b=f(x_data, y_data)
print(a,b)
```

```python
plt.plot(x_data, y_data, "bo", label="Points") # On affiche le nuage de points données par x_data et y_data

y_data = a*np.sin(0.5*np.pi*x_data) + b*np.cos(0.5*np.pi*x_data) # On défnit la fonction de départ avec la valeur de a et de b trouvé

plt.plot(x_data,y_data, "r-", label="Courbe") 
plt.xlim(-10, 10) 
plt.legend() 
plt.title("Courbe de la meilleur approximation et un nuage des points") 
plt.show()
```
